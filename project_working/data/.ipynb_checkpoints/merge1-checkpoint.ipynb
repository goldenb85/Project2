```python
import pandas as pd
import geopandas as gpd
```

```python
df = pd.read_json('quartileData.json')
```

```python
df.head()
```

```python
df2 = gpd.read_file('stateshape.json')
# df2 = geopandas.read_json('stateshape.json')
```

```python
df2.head()
```

```python
new_df = df2[['NAME', 'geometry']]
new_df.head()
```

```python
new_df = new_df.rename(columns={'NAME': 'State'})
new_df.head()
```

```python
combined_df = new_df.merge(df, on='State')
combined_df.head()
```

```python
import fiona; fiona.supported_drivers
```

```python
combined_df.to_file("combined_df.geojson", driver='GeoJSON')
```
