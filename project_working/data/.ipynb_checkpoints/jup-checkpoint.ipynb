```python
import pandas as pd
```

```python
df = pd.read_json('population.json')
```

```python
df.head()
```

```python
df['State'] = df.data.map(lambda i: i['State'])
```

```python
df['Pop'] =  df.data.map(lambda i: i['Population'])
```

```python
df['Year'] = df.data.map(lambda i: i['Year'])
```

```python
df.head()
```

```python
dfUSA = df[['State',"Pop","Year"]]
```

```python
dfUSA
```

```python
dfUSA.head()
```

```python
dfUSA.State.head()
```

```python
# dfUSA.to_dict(orient="records")
```

```python
new_df = dfUSA[dfUSA['Year'] == '2018']
```

```python
new_df
```

```python
us_pop_2018 = '327200000'
```

```python
new_df['pop/uspop'] = dfUSA['Pop'] / 327200000
```

```python
new_df.head()
```

```python
new_df['per_US'] = new_df['pop/uspop'] * 100
```

```python
new_df.head()
```

```python
quartile_data = new_df[['State', 'Pop', 'Year', 'per_US']]
```

```python
quartile_data.head()
```

```python
# the qcut
import numpy as np
import seaborn as sns
```

```python
bin_labels = ['4th', '3rd', '2nd', '1st']
```

```python
quartile_data['quantile'] = pd.qcut(quartile_data['per_US'], q=4 , labels=bin_labels)
```

```python
quartile_data
```

```python
quartile_data.to_dict(orient="records")
```

```python
quartile_data.to_json('quartileData.json', orient='records')
```
