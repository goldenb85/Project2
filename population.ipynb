```python
file_name = "Resources/population.json"
import pandas as pd
df = pd.read_json(file_name)
```

```python
from pandas.io.json import json_normalize
df=json_normalize(df['data']) 
```

```python
df.head()
```

```python
df = df.reset_index()


```

```python
df
```

```python
df=df.rename(columns={"index":"id_num","ID State": "id_state", "State": "state",
                   "ID Year":"id_year","Year":"year",
                   "Population":"population","Slug State":"slug_state"})
```

```python
df["id_num"]=df["id_num"]+1
```

```python
df
```

```python
engine = create_engine('postgres+psycopg2://postgres:password@localhost:5432/project2_db')
```

```python
df.to_sql(name='population_table',con=engine,if_exists='replace',index=False)
```
